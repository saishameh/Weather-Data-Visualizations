```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\sdais\Downloads\Daily Weather.csv")
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'line', title = 'Weather', xlabel = 'Daily Weather', ylabel = 'Weather Data')
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Humidity', y = 'Wind Speed', s = 500, c = 'Blue')
```

```python
df.plot.hist(bins = 10)
```

```python
df.plot.pie(y='Temperature',figsize=(8,18))
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(df.index, df['Temperature'], color='blue', marker='o')
axs[0, 0].set_title('Temperature Over Time')

axs[0, 1].bar(df.index, df['Humidity'], color='green')
axs[0, 1].set_title('Humidity Over Time')

axs[1, 0].scatter(df.index, df['Wind Speed'], color='orange')
axs[1, 0].set_title('Wind Speed Over Time')

axs[1, 1].bar(df.index, df['Precipitation'], color='purple')
axs[1, 1].set_title('Precipitation Over Time')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Temperature'])
plt.title('Temperature Distribution')
plt.show()
```
